```python
import os
from pathlib import Path

def add_docx_to_extensionless_files(folder: Path) -> None:
    for item in folder.iterdir():
        if item.is_file():
            # Skip hidden files like ".bashrc" (name starts with dot) and files that already have an extension
            if item.name.startswith('.') or item.suffix:
                continue
            new_path = item.with_name(item.name + '.docx')
            # Avoid overwriting existing files
            if new_path.exists():
                print(f"Skipping (target exists): {new_path}")
                continue
            item.rename(new_path)
            print(f"Renamed: {item.name} -> {new_path.name}")
```

```python
downloads = Path("downloads")
add_docx_to_extensionless_files(downloads)
```
