````python
import re
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage
import os
from dotenv import load_dotenv
load_dotenv()
````

````python
address = "https://litellm.labs.jb.gg/"
api_key = os.getenv("LITELLM_KEY")
model_name = "gpt-5-nano"
````

````python
batch_size = 10
system_prompt = "You are a helpful assistant that knows both German and Russian. Helpful translator"
base_prompt = '''
You're given a list of German words or short phrases, separated by a new line after the indicator WORDS_LIST. For each word generate a json {"word": word, "translation": translation of the word to Russian, "example": example of the word usage, "example_translation": translation of the example to russian}. Generated examples should not be templated; they should be natural phrases taken from the natural texts. Each JSON should start on a new line, forming a JSONL file. Start list of JSON dicts with indicator TRANSLATION_LIST```, then list translations of words in the list, end with indicator ```\n\nWORDS_LIST:\n
'''
````

````python
word_list_file = "C:/Timur/GoogleDrive/Timur/Deutsch/words_B1.txt"
````

````python
with open(word_list_file, "r", encoding="utf-8") as f:
    word_list = [line.strip() for line in f if line.strip()]
word_list = list(set([re.sub(r" {2,}", " ", word) for word in word_list]))
words_batches = [word_list[i:i+batch_size] for i in range(0, len(word_list), batch_size)]
````

````python
prompts = []
messages = []
for batch in words_batches:
    batch_str = "\n".join(batch)
    prompt = (base_prompt + batch_str).strip()
    message = [SystemMessage(content=system_prompt), HumanMessage(content=prompt)]
    prompts.append(prompt)
    messages.append(message)
````

````python
llm = ChatOpenAI(base_url=address, api_key=api_key, model=model_name)
````

````python
short_mes = messages[:2]
````

````python
response = llm.batch(short_mes)
````
