```python
import json
from generate_vocab import write_vocab_csv
```

```python
vocab_file = "vocab_cards_gpt5-nano.jsonl"
```

```python
records = []
with open(vocab_file, "r", encoding="utf-8") as f:
    for line in f:
        if line.strip():  # skip blank lines
            records.append(json.loads(line))
```

```python
write_vocab_csv(records, "vocab_cards_gpt5-nano.csv")
```
